# **ANALYSIS BASED ON SHIPPING MODE**

```python
import matplotlib.pyplot as plt
gds = df.groupby('Ship Mode').sum(numeric_only=True)
gds[['Sales', 'Profit']].plot(kind='bar')
plt.xlabel('Ship Mode')
plt.ylabel('Amount')
plt.title('Ship Mode-Wise Analysis:Total Sales,Total Profit')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

 From The Above It Is Clear That More The Number Of ***Sales*** More Will Be The **Profit**


---


*  We Can Observe More profit For Standard Class Shipping Mode
and Less profits for Same Day Shipping Mode.
* The Second Class Shipping Mode stands as the Second highest Profitable Shipping Mode
* For the shippings modes First class ,same day and second class
by analyzing and optimizing these processes(order fulfillment, packaging, and handling), you can potentially reduce costs and improve efficiency, thereby increasing profitability.

```python
gd=df.groupby('Ship Mode')
colors = {'First Class': 'red', 'Second Class': 'blue', 'Same Day': 'green', 'Standard Class': 'orange'}
for mode ,data in gd:
  plt.scatter(data["Sales"],data["Profit"],label=mode,color=colors[mode])
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.title("Sales vs profit by shipping mode")
plt.legend()
plt.show()
```

# **conclusions**:


From the above plotted graph between sales and profits for every shipping mode we can conclude the following::


** comparitively we can observe low profits even for high sales(In Scatter Plot).
we have to look into the following things for understanding the trends between sales and profits::




 1.Shipping Costs: Due to high shipping costs we might be facing the low profits situation even for high sales

2.price fixation and discounts : May be we must change the discounts
according to the season ,that might increase our profits


3. By analyzing and optimizing these processes(order fulfillment, packaging, and handling), you can potentially reduce costs and improve efficiency, thereby increasing profitability.
